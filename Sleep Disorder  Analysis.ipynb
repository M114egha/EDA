### WHAT IS SLEEP DISORDER?

- SLEEP DISORDER is a condition that makes it hard for a person to sleep properly.
- Insomnia : Trouble falling or staying asleep.
- Sleep Apnea :  Due to airway blockage person's breathing stops and starts repeatedly during sleep.
- Narcolepsy : Sudden sleep attacks during the day.


### WHY SLEEP DISORDER ANALYIS IS IMPORTANT?

 Sleep disorder analysis is crucial because poor sleep affects both physical and mental health. It can lead to conditions like heart disease, high blood pressure, obesity, and increased stress levels. Lack of quality sleep also reduces focus, productivity, and overall well-being. By analyzing factors like stress, sleep quality, and lifestyle habits, early detection and proper treatment of sleep disorders become possible, improving long-term health and daily performance.

### DATASET OVERVIEW

Dataset Columns:

Person ID: An identifier for each individual.

Gender: The gender of the person (Male/Female).

Age: The age of the person in years.

Occupation: The occupation or profession of the person.

Sleep Duration (hours): The number of hours the person sleeps per day.

Quality of Sleep (scale: 1-10): A subjective rating of the quality of sleep, ranging from 1 to 10.

Physical Activity Level (minutes/day): The number of minutes the person engages in physical activity daily.

Stress Level (scale: 1-10): A subjective rating of the stress level experienced by the person, ranging from 1 to 10.

BMI Category: The BMI category of the person (e.g., Underweight, Normal, Overweight).

Blood Pressure (systolic/diastolic): The blood pressure measurement of the person, indicated as systolic pressure over diastolic 
pressure.

Heart Rate (bpm): The resting heart rate of the person in beats per minute.

Daily Steps: The number of steps the person takes per day.

Sleep Disorder: The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df["BMI Category"] = df["BMI Category"].replace({
    "Normal Weight": "Normal"
})
```

```python
plt.figure(figsize=(8, 6))
sleep_disorder_counts = df['Sleep Disorder'].value_counts()
plt.pie(sleep_disorder_counts, labels=sleep_disorder_counts.index, autopct='%.1f%%', colors=['red', 'orange', 'green'])
plt.title('Percentage of People with Sleep Disorders')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df , x='Sleep Disorder' , hue='Gender')
plt.title('Distribution of Sleep disorder by Gender')
plt.show()
```

INFERENCE:
- Males in comparison with females  have no sleep disorder
- Sleep Apnea is found more in females
- Insomnia affects both male and female

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Age'], y=df['Stress Level'], hue=df['Sleep Disorder'])
plt.title("Relationship Between Age, Stress Level, and Sleep Disorders")
plt.show()
```

INFERENCE:
- People with stress level above 6 have Insomnia and Sleep Apnea
- People with stress level less than 4 have no disorder
- Insomnia is spread across all ages 
- Sleep Apnea is found  for age higher than 40

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Heart Rate", hue="Sleep Disorder", kde=True)
plt.title("Distribution of Heart Rate across sleep Disorders")
plt.show()
```

INFERENCE:
- People with no disorder have lower heart rates
- Sleep Apnea have the highest heart rates
- Insomnia shows some variation around 65-75 range

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, y='Daily Steps',x='BMI Category', hue='Sleep Disorder')
plt.title("Relationship Between BMI, Physical Activity, and Sleep Disorders")
plt.show()
```

INFERENCE:

Overweight BMI Category:
- People with sleep apnea have the highest daily steps among overweight individuals.

Obese BMI Category:
- Obese individuals have the lowest daily steps across all BMI categories.

Normal BMI Category:
- Individuals with no sleep disorder tend to have the highest daily steps

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[['Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 
                'Stress Level', 'BMI Category', 'Heart Rate', 'Daily Steps']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap ')
plt.show()
```

INFERENCE:
- Sleep Duration & Quality of Sleep are positively correlated ensuring more sleep durationis strongly associated with better sleep quality
- Daily Steps & Physical Activity Level :Higher daily steps indicate a more active lifestyle.
- Stress Level & Heart Rate : Higher stress levels are linked to increased heart rate.

```python
occupation_counts = df['Occupation'].value_counts().reset_index()
occupation_counts.columns = ['Occupation', 'Count']
sorted_occupations = occupation_counts.sort_values(by='Count', ascending=True)['Occupation']
plt.figure(figsize=(8, 7))
sns.countplot(data=df, y='Occupation', hue='Sleep Disorder', order=sorted_occupations)
plt.title("Sleep Disorders by Occupation")
plt.show()
```

INFERENCE:
- Nurses have the highest count of individuals experiencing Sleep Apnea and Insomnia. 
- Teachers and Salespersons show high count of Insomnia, which could be linked to stress from workload and job pressure.

```python
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/', expand=True)
df['Systolic'] = pd.to_numeric(df['Systolic'])
df['Diastolic'] = pd.to_numeric(df['Diastolic'])
```

sys_bins (Systolic Blood Pressure Categories)
- 0 - 120 → "Normal"  (Healthy range)
- 120 - 140 → "Elevated"   (Borderline high blood pressure)
- 140+ (np.inf) → "Hypertension"   (High blood pressure)

dia_bins (Diastolic Blood Pressure Categories)
- 0 - 80 → "Normal"
- 80 - 90 → "Elevated"
- 90+ (np.inf) → "Hypertension"

```python
labels = ["Normal", "Elevated", "Hypertension"]
sys_bins = [0, 120, 140, np.inf]
dia_bins = [0, 80, 90, np.inf]

df['Systolic Category'] = pd.cut(df["Systolic"], labels=labels, bins=sys_bins)
df['Diastolic Category'] = pd.cut(df["Diastolic"], labels=labels, bins=dia_bins)
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Systolic Category", hue="Sleep Disorder", data=df)
plt.title("Systolic Blood Pressure vs Sleep Disorder")
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(x="Diastolic Category", hue="Sleep Disorder", data=df)
plt.title("Diastolic Blood Pressure vs Sleep Disorder")
plt.show()
```

INFERENCE:
- People with normal blood pressure (both systolic & diastolic) mostly do not have sleep disorders.
- Insomnia is common in individuals with elevated systolic and diastolic BP.
- Hypertension (both systolic & diastolic) is strongly associated with sleep disorders:
  Sleep apnea is the most common disorder among hypertensive individuals.
- Overall, higher blood pressure (both systolic and diastolic) correlates with an increased risk of sleep disorders.

CONCLUSION:
- Stress level impacts sleep. Higher stress levels are associated with poor sleep quality.
- Sleep quality is crucial. Poor sleep quality is seen in individuals with sleep apnea and insomnia.
- Blood pressure and sleep disorders are linked. High BP increases the risk of sleep apnea and insomnia.
- Occupation influences sleep quality. High-stress jobs like nursing show more sleep disorders.
- Obesity & inactivity worsen sleep issues. Insomnia is more common among less active individuals.

Prescriptions for Sleep Disorder Management:

For Sleep Apnea:
- Use a special breathing mask (CPAP) while sleeping.
- Lose weight if overweight to help with breathing.
- Avoid alcohol and sleeping pills.

For Insomnia:
- Follow a consistent sleep schedule (same bedtime & wake-up time).
- Reduce screen time & caffeine before bed.
- Practice relaxation techniques like meditation or deep breathing.

- Regularly monitor blood pressure and consult a doctor.
- Manage stress through yoga, meditation, or therapy.
- Reduce screen time & caffeine before bed.
- Address mental health issues like anxiety & depression with professional help.
